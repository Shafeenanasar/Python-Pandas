```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='/content/supply_chain_data.csv'
```

```python
df=pd.read_csv(path)
df
```

```python
df['Product type'].unique()
```

```python
df['Product type'].value_counts()
```

# No: of products sold in each category


```python
sales=df.groupby('Product type')['Number of products sold'].sum().sort_values(ascending=False)
sales
```

```python
plt.title('Products sold in each category')
plt.bar(sales.index,sales.values,color='red')
plt.xlabel('Products')
plt.ylabel('sales')
```

```python
sale=sns.barplot(x='Product type',y='Number of products sold',data=df,palette='rocket',errorbar=None)
plt.title('ANALYSIS ON PRODUCTS AND THEIR RESPECTIVE SALES')
plt.xlabel('Product segment')
plt.ylabel('Units Sold')
```

# HISTOGRAM

```python
plt.hist(df['Shipping times'],color="orange",edgecolor='red')
plt.title("ANALYSIS ON SHIPPING TIME DISTRIBUTION")
plt.xlabel("Shipping time")
plt.ylabel("Frequency")
```

```python
customer_demographic=sns.barplot(x='Customer demographics',y='Order quantities',data=df,palette='viridis',errorbar=None)
plt.title('ANALYSIS ON HAIRCARE PRODUCTS CUSTOMER DEMOGRAPHICS BASED ON QUANTITY')
plt.xlabel('Demographic')
plt.ylabel('Quantities')
```

```python
plt.figure(figsize=(10,10))
inspectionc=sns.barplot(x='Product type',y='Availability',data=df,hue='Inspection results',palette='cubehelix',errorbar=None)
plt.title('DETAILED ANALYSIS ON PRODUCTS AND AVAILABLITY ALONG WITH THEIR INSPECTION STATUS.')
plt.xlabel('Product type')
plt.ylabel('quantity')
```

```python
plt.figure(figsize=(10,10))
ax=sns.barplot(x='Product type',y='Lead time',hue='Transportation modes',data=df,palette='magma')
plt.title('AVERAGE LEAD TIME BY TRANSPORTATION MODE')
plt.xlabel('Product Segment')
plt.ylabel('Lead time')
```
