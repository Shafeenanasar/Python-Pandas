**USing apply function map the data according to fare**

```python
def fare_category(x):
  if x<20:
    return ("Low Fare")
  elif x>20 and x<100:
    return ("Medium fare")
  else:
    return ('High fare')

```

```python
df['Ticket category']=df['Fare'].apply(fare_category)
```
