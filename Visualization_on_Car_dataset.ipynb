YEAR WISE REPORT GENERATE

```python
year_wise_sales=df['year'].value_counts()
year_wise_sales
```

Ploting graph-- year wise sales

labels

x:year
y:sales

```python
plt.plot(year_wise_sales.index,year_wise_sales.values,color='red',marker='*')
plt.xlabel('year')
plt.ylabel('sales')
plt.title('YEAR WISE SALES REPORT')
```

**PLOT Graphical representation between sales and year**

The highest number of sales is occured in the year of 2011 with the sales of 3586

The below bar chart is also a representation between the sales and year

```python
plt.bar(year_wise_sales.index,year_wise_sales.values,color='red')
plt.xlabel('year')
plt.ylabel('sales')
plt.title('YEAR WISE SALES REPORT')
```





>**BRAND WISE TOP 10 SALES REPORT**

>**Filtering the vehicle based on make**



```python
kia_filter=df[df['make']=='kia']
kia_filter
```

```python
kia_filter['model'].unique()
```

```python
BMW_filter=df[df['make']=='BMW']
BMW_filter
```

```python
variant=BMW_filter.groupby('model')['sellingprice'].mean()
variant
```

```python
plt.bar(variant.index,variant.values,color='red',edgecolor='white')
plt.title('Avg selling price of BMW Model')
plt.xlabel('Model')
plt.ylabel('Avg selling price')
plt.xticks(rotation=90)
```

The above barplot illustrates the average selling price for BMW variants with the model ***M4*** having the ***highest*** average selling price and **least selling prices** of the model **z3**

HIGHEST MARKET PRICE

The bar graphical representation gives the maximum market price of a vehicle

```python
top_market_price=df.groupby('model')['mmr'].max().sort_values(ascending=False).head(10).reset_index()
top_market_price
```

```python
plt.bar(top_market_price['model'],top_market_price['mmr'],color='red')
plt.title('Highest market price vehicles')
plt.xlabel('Model')
plt.ylabel('Price')
plt.xticks(rotation=90)
```

The below bar graphical representation gives the total price of all the makes of the vehicle

```python
top_market_price=df.groupby('make')['mmr'].sum().sort_values(ascending=False).head(10).reset_index()
top_market_price
```

```python
plt.bar(top_market_price['make'],top_market_price['mmr'],color='red')
plt.title('Highest market price vehicles')
plt.xlabel('Make')
plt.ylabel('Price')
plt.xticks(rotation=90)
```
